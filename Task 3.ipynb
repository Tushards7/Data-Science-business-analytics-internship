## EDA - 'Exploratory Data Analysis'

```python
# describing the dataset with help of seaborn pairplot
sns.pairplot(df)
```

```python
# Finding correlation between the variables and vizualizing with the help of heatmap
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
import plotly.graph_objects as go

labels = ['West', 'East', 'Central', 'South']
values = [3203, 2848, 2323, 1620]

fig = go.Figure(data=[go.Pie(labels=labels, values=values, textinfo='label+percent',
                             insidetextorientation='radial'
                            )])
fig.show()
```

```python
plt.figure(figsize=(10,8))
df.groupby('Category')['Profit','Sales'].sum().plot.bar()
plt.ylabel('Profit/Sales',fontsize=14)
plt.xlabel('Category',fontsize=14)
plt.title('Categorywise Sales vs profit',fontsize=15)
plt.show
```

```python
Category_d = df.groupby('Category').Discount.sum().nlargest(n=100)

Category_d.plot(kind='bar',figsize=(10,5),fontsize=14)
plt.ylabel('Discount',fontsize=14)
plt.xlabel('Category',fontsize=14)
plt.title('Discount availed on different categories',fontsize=15)
plt.show()
```

#### Technology is granted least discount as compared to other categories

```python
category_s = df.groupby('Category').Sales.sum().nlargest(n=100)
category_p = df.groupby('Category').Profit.sum().nlargest(n=100)

category_s.plot(kind='bar',figsize=(10,5),fontsize=14)
category_p.plot(kind='bar',figsize=(10,5),fontsize=14,color='red')
plt.xlabel('Category',fontsize=15)
plt.ylabel('Total Sales/profit',fontsize=15)
plt.title('Top Category Sales vs profit',fontsize=15)
plt.show()
```

```python
Sub_category_s = df.groupby('Sub-Category').Sales.sum().nlargest(n=100)
Sub_category_p = df.groupby('Sub-Category').Profit.sum().nlargest(n=100)

Sub_category_s.plot(kind='bar',figsize=(10,5),fontsize=14)
Sub_category_p.plot(kind='bar',figsize=(10,5),fontsize=14,color='red')
plt.xlabel('Sub-category',fontsize=15)
plt.ylabel('Total Sales/Profit',fontsize=15)
plt.title('Top Sub-Category Sales vs Profit',fontsize=15)
plt.show()
```

```python
sub_cat_p = df.groupby('Sub-Category').Profit.sum().nlargest(n=100)
sub_cat_p.plot(kind='bar',figsize=(10,5),fontsize=14)
plt.xlabel('Sub-Category',fontsize=14)
plt.ylabel('Profit',fontsize=14)
plt.show()
```

#### The above graph clearly shows that Copiers and Phones have the highest sales and profit whereas tables has negative profit

```python
Segment_s = df.groupby('Segment').Sales.sum().nlargest(n=100)
Segment_p = df.groupby('Segment').Profit.sum().nlargest(n=100)

Segment_s.plot(kind='bar',figsize=(10,5),fontsize=14)
Segment_p.plot(kind='bar',figsize=(10,5),fontsize=14,color='red')
plt.xlabel('Segment',fontsize=14)
plt.ylabel('Total Sales/Profit',fontsize=14)
plt.title('Segment Sales vs Profit',fontsize=15)
plt.show()
```

#### Consumer segment has the maximum sales and profits whereas home office segment has the minimum

```python
df.groupby('Segment').Discount.sum().plot(kind='bar',figsize=(10,5),fontsize=14)
plt.xlabel('Segments',fontsize=14)
plt.ylabel('Discount',fontsize=14)
plt.title('Discount on differnt Segments',fontsize=15)
plt.show()
```

#### The above graph shows that consumer segment is granted maximum discount as compred to other segments

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='Discount',y='Profit',data=df)
```

```python
plt.figure(figsize = (6,6))
textprops = {"fontsize":15}
plt.title('Category')
plt.pie(df['Segment'].value_counts(), labels=df['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

#### This chart shows that office supplies has the highest sales and profit

```python
State_s = df.groupby('State').Sales.sum().nlargest(n=15)
State_p = df.groupby('State').Profit.sum().nlargest(n=15)

State_s.plot(kind='bar',figsize=(10,5),fontsize=14)
State_p.plot(kind='bar',figsize=(10,5),fontsize=14,color='red')
plt.xlabel('State',fontsize=14)
plt.ylabel('Total sales/profit',fontsize=14)
plt.title('Top 15 statewise Sales vs Profit',fontsize=15)
plt.show()
```

```python
df.plot(kind = "scatter",figsize = (10,5), x = "Sales", y= "Profit", c = "Discount",grid='True',fontsize = 16, colormap = "viridis")
plt.ylabel('Total Profits',fontsize = 16)
plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 16)
```
